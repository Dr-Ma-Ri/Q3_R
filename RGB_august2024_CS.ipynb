```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Load the data
file_path = 'RGB082024_2.csv'  # Adjust the path if necessary
data = pd.read_csv(file_path)

# Define the variables of interest
variables = ['VEG', 'GLI', 'HUE', 'SHP', 'NGRDI', 'VARI', 'SRPI']

# Perform two-way ANOVA for each variable
for var in variables:
    # Define the model formula
    model_formula = f'{var} ~ C(Genotype) + C(Field) + C(Genotype):C(Field)'

    # Fit the model
    model = ols(model_formula, data=data).fit()

    # Perform ANOVA
    anova_table = anova_lm(model, typ=2)

    # Print the ANOVA table
    print(f"ANOVA Table for {var}:")
    print(anova_table)
    print("\n")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
file_path = 'RGB082024_2.csv'  # Adjust the path if necessary
data = pd.read_csv(file_path)

# Select the relevant columns
variables = ['VEG', 'GLI', 'NGRDI', 'VARI', 'SRPI']
selected_data = data[['Field', 'Genotype'] + variables]

# Pivot the data to create a matrix for each variable using pivot_table
# and aggregate duplicates using the mean (you can change the aggfunc)
pivoted_data = {}
for var in variables:
    pivoted_data[var] = selected_data.pivot_table(
        index='Genotype',
        columns='Field',
        values=var,
        aggfunc='mean'  # Use 'mean' to average duplicate values
    )

# Create a figure with subplots
fig, axs = plt.subplots(nrows=len(variables), figsize=(12, 6 * len(variables)))

# Iterate over the variables and create heatmaps
for i, var in enumerate(variables):
    sns.heatmap(pivoted_data[var], annot=True, cmap='coolwarm', square=True, fmt='.2f', ax=axs[i])
    axs[i].set_title(f'{var} Index by Genotype and Field')
    axs[i].set_xlabel('Field')
    axs[i].set_ylabel('Genotype')

# Layout so plots do not overlap
fig.tight_layout()

plt.show()

```

```python
import pandas as pd
from scipy.stats import f_oneway

# Load the data
file_path = 'RGB082024_2.csv'  # Adjust the path if necessary
data = pd.read_csv(file_path)

# Select the relevant columns
variables = ['VEG', 'GLI', 'HUE', 'SHP', 'NGRDI', 'VARI', 'SRPI']
selected_data = data[['Field', 'Genotype'] + variables]

# Calculate mean values for each field and genotype
mean_values_field = selected_data.groupby('Field')[variables].mean()
mean_values_genotype = selected_data.groupby('Genotype')[variables].mean()

# Print the mean values
print("Mean Values by Field:")
print(mean_values_field)
print("\nMean Values by Genotype:")
print(mean_values_genotype)

# Identify fields with higher mean values for each variable
for var in variables:
    max_field = mean_values_field[var].idxmax()
    max_field_value = mean_values_field[var].max()
    print(f"Field with highest {var}: {max_field} with value {max_field_value}")

# Identify genotypes with higher mean values for each variable
for var in variables:
    max_genotype = mean_values_genotype[var].idxmax()
    max_genotype_value = mean_values_genotype[var].max()
    print(f"Genotype with highest {var}: {max_genotype} with value {max_genotype_value}")

# Perform ANOVA and interpret results
for var in variables:
    print(f"ANOVA Results for {var}:")
    # Example ANOVA using f_oneway (you can use your existing ANOVA function)
    f_stat, p_value = f_oneway(*[selected_data[selected_data['Genotype'] == genotype][var] for genotype in selected_data['Genotype'].unique()])
    if p_value < 0.05:
        print(f"Significant difference among genotypes for {var} (p-value: {p_value})")
    else:
        print(f"No significant difference among genotypes for {var} (p-value: {p_value})")

    # Perform ANOVA for fields
    f_stat, p_value = f_oneway(*[selected_data[selected_data['Field'] == field][var] for field in selected_data['Field'].unique()])
    if p_value < 0.05:
        print(f"Significant difference among fields for {var} (p-value: {p_value})")
    else:
        print(f"No significant difference among fields for {var} (p-value: {p_value})")
    print()
```
